### Comparison of C++, FCCAnalysis and Python standalone streering scripts' time output (each version was run from Random Access Memory)

```python
import matplotlib.pyplot as plt
from tools.utilities  import json2dict_py, plot_graph, divide_by_threads, json2dict_cpp_updated, json2dict_py_updated, json2dict_fcc_updated
```

```python
data_fcc_wocr = json2dict_fcc_updated("/home/ihor/PycharmProjects/fccperf/json-RDF-data/perf_fccanalysis_wocr.json", [1, 65, 3])
data_py_wocr = json2dict_py_updated("/home/ihor/PycharmProjects/fccperf/json-RDF-data/perf_py_wocr.json", [1, 65, 3])
data_cpp_wocr = json2dict_cpp_updated("/home/ihor/PycharmProjects/fccperf/json-RDF-data/perf_cpp_wocr.json", [1, 65, 3])
```

```python
elapsed_fcc_wocr = data_fcc_wocr["elapsed"]
elapsed_py_wocr = data_py_wocr["elapsed"]
elapsed_cpp_wocr = data_cpp_wocr["elapsed"]
n_threads = [i for i in range(1, 65, 3)]
# n_threads_1 = [i for i in range(1, 65, 1)]

ratio_py = [abs(py - fcc)/fcc for py, fcc in zip(elapsed_py_wocr, elapsed_fcc_wocr)]
ratio_cpp = [abs(cpp - fcc)/fcc for cpp, fcc in zip(elapsed_cpp_wocr, elapsed_fcc_wocr)]
```

```python
ratio_cpp
```

```python
plt.figure(figsize=(16, 10), facecolor='white')
ax = plt.axes()
ax.set_facecolor('white')
ax.set_alpha(0.1)
plt.plot(n_threads, ratio_py, color='b', label='Elapsed Time Ratio Python wocr', linewidth=3)
plt.vlines(x = 32, ymin = 75, ymax = 250, color = 'r', label = '32 threads', linestyles="dashed")
plt.xlabel("Number of threads")
plt.ylabel("Elapsed time Ratio")
plt.title("")
plt.legend()
plt.xlim((0, 65))
plt.ylim((0.5, 0.8))
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16, 10), facecolor='white')
ax = plt.axes()
ax.set_facecolor('white')
ax.set_alpha(0.1)
plt.plot(n_threads, elapsed_py_wocr, color='b', label='Elapsed Time Py wocr', linewidth=3)
plt.plot(n_threads, elapsed_fcc_wocr, color='g', label='Elapsed Time FCC wocr', linewidth=3)
plt.vlines(x = 32, ymin = 75, ymax = 250, color = 'r', label = '32 threads', linestyles="dashed")
plt.xlabel("Number of threads")
plt.ylabel("Elapsed time")
plt.title("")
plt.legend()
plt.xlim((0, 65))
plt.ylim((0, 355))
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16, 10), facecolor='white')
ax = plt.axes()
ax.set_facecolor('white')
ax.set_alpha(0.1)
plt.plot(n_threads, ratio_cpp, color='b', label='Elapsed Time Ratio C++ wocr', linewidth=3)
plt.vlines(x = 32, ymin = 75, ymax = 250, color = 'r', label = '32 threads', linestyles="dashed")
plt.xlabel("Number of threads")
plt.ylabel("Elapsed time Ratio")
plt.title("")
plt.legend()
plt.xlim((0, 65))
plt.ylim((0.4, 0.9))
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16, 10), facecolor='white')
ax = plt.axes()
ax.set_facecolor('white')
ax.set_alpha(0.1)
plt.plot(n_threads, elapsed_cpp_wocr, color='b', label='Elapsed Time C++ wocr', linewidth=3)
plt.plot(n_threads, elapsed_fcc_wocr, color='g', label='Elapsed Time FCC wocr', linewidth=3)
plt.vlines(x = 32, ymin = 75, ymax = 250, color = 'r', label = '32 threads', linestyles="dashed")
plt.xlabel("Number of threads")
plt.ylabel("Elapsed time")
plt.title("")
plt.legend()
plt.xlim((0, 65))
plt.ylim((0, 250))
plt.grid()
plt.show()
```
