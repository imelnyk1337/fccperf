```python
import matplotlib.pyplot as plt
import json
```

```python
print(plt)
print(json)
```

```python
bench_time = list()
nevents = list()
real_time = list()
user_time = list()
syst_time = list()
with open("perf_cpp.json", "r") as perf_js:
    data = json.load(perf_js)
    print(data["2"]["time"])
    for i in range(1, 65):
        bench_time.append(float(data[str(i)]["time"]))
        nevents.append(float(data[str(i)]["nevents"]))
        real_time.append(float(data[str(i)]["real"]))
        user_time.append(float(data[str(i)]["user"]))
        syst_time.append(float(data[str(i)]["sys"]))
```

```python
real_time
# 13m325s -> 13.
```

```python
def plot_graph(data: list, ylims: [], title: str):
    plt.figure(figsize=(12, 8), facecolor='white')
    ax = plt.axes()
    ax.set_facecolor('white')
    ax.set_alpha(0.1)
    plt.grid()
    plt.xlim([0, 65])
    plt.ylim(ylims)
    plt.title("", fontsize=14)
    plt.plot([x for x in range(1, 65)], data, linewidth=2.5, color="red", label=title)
    plt.legend(fontsize=12)
    plt.xlabel("Number of CPUs", fontsize=12)
    plt.ylabel(title, fontsize=18)
    plt.show()
    plt.savefig("time_performance")
```

```python
plot_graph(bench_time, [930, 1500], "time")
```

```python
plot_graph(real_time, [930, 1500], "real")
```

```python
plot_graph(user_time, [900, 60_000], "user")
```

```python
plot_graph(syst_time, [10, 500], "sys")
```

```python
plot_graph(nevents, [40_000, 63_632], "nevents")
```
