```python
import matplotlib.pyplot as plt
import json
from tools.utilities  import json2dict_cpp, plot_graph, divide_by_threads, compute_nevents, json2dict
```

```python
data_python_wo_cache = json2dict("/home/ihor/PycharmProjects/fccperf/jsonfiles/perf_python_wo_cache_1_64_1.json", [1, 65, 1])
data_cpp_wo_cache = json2dict_cpp("/home/ihor/PycharmProjects/fccperf/jsonfiles/perf_cpp_without_cache.json", [1, 65, 1])
n_threads = [i for i in range(1, 65, 1)]
```

```python
plt.figure(figsize=(16, 10), facecolor='white')
ax = plt.axes()
ax.set_facecolor('white')
ax.set_alpha(0.1)
plt.plot(n_threads, data_python_wo_cache["real_time"], color='b', label='Real time (Python)', linewidth=3)
plt.plot(n_threads, data_cpp_wo_cache["real_time"], color='g', label='Real time (C++)', linewidth=3)
# plt.plot(n_threads, real_time_c_wo_cache, color='y', label='Real time (C++) cache realise', linewidth=3)
plt.vlines(x = 32, ymin = 75, ymax = 250, color = 'r', label = '32 threads', linestyles="dashed")
plt.xlabel("Number of threads")
plt.ylabel("Real time")
plt.title("Comparison of C++ and Python versions")
plt.legend()
plt.xlim((0, 65))
plt.ylim((0, 1350))
plt.grid()
plt.show()
```
